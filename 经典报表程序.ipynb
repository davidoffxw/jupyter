```python
import pandas as pd
import os

path='/users/Davidoff/desktop/每日报表/'
excelfile='10094-20210101-20210125.xls'

# output_day='2021-1-12'

 
#df=pd.DataFrame(pd.read_excel(path+excelfile))
df_original=pd.read_excel(path+excelfile,sheet_name='1',header=1)

df=df_original[['保单号','投保人名称','车牌号','签单保费（原币）','承保确认日期','险种名称','保险起期','车辆种类名称','业务员名称']]

agent_name=[ '臧衍娟','牛一凡'] 

df_agent_name=df[df['业务员名称'].isin(agent_name) ]

df_agent_name['承保确认日期2']=pd.to_datetime(df_agent_name['承保确认日期']) 

df_agent_name = df_agent_name.set_index('承保确认日期2')

def output_daytofile(output_day):
    
    # output_day = output_day

    df_agent_name_output=df_agent_name.loc[output_day]
    
    df_agent_name_output.to_excel(path+'经典'+output_day+'.xls',index = None)

#print(df_agent_name_output)
#print(df_agent_name_output['签单保费（原币）'].sum())

def writetofile():
    df_agent_name.to_excel(path+'经典'+excelfile,index=None)
    
    
```
